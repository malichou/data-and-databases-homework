```python
driver.get('https://kinimaallagis.gr/category/omilies/')
```

```python
items = driver.find_elements_by_class_name("the-title")
for item in items:
    print('---------------------------------------')
    print(item.text)
```

```python
items = driver.find_elements_by_class_name('item-wrapper')
for item in items:
    print('------------------------------------')
    print(item.text) 
```

```python
dates = driver.find_elements_by_tag_name('time')
for date in dates:
    print('------------------------------------')
    print(date.text) 
```

```python
items = driver.find_elements_by_class_name('item-wrapper')

rows = []
for item in items:
    row={}
    row['Title'] = item.find_element_by_class_name('the-title').text
    row['Date'] = item.find_element_by_tag_name('time').text
    row['URL'] = item.find_element_by_tag_name('a').get_attribute('href')
    rows.append(row)

df1 = pd.DataFrame(rows)
df1.head()
```
